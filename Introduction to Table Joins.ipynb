```python
# inner join
names.merge(jobs, how = "inner", on = "iid")
```

```python
# Duplicate Inner Join
jobs_with_dups = pd.read_csv(r"C:\Users\oriha\Desktop\Serious Sql Excercises\csv\jobs_with_duplicates.csv")
names.merge(jobs_with_dups, how = "inner")
```

```python
# Duplicate Left Join
names.merge(jobs_with_dups, how = "left")
```

```python
# Left Semi Join
in_both = names['iid'].isin(jobs_with_dups['iid'])
names[in_both][["iid","first_name"]]
```

```python
# Anti join
in_both = names['iid'].isin(jobs_with_dups['iid']) == False
names[in_both][["iid","first_name"]]
```

```python
# Joining on Null Values
names_with_nulls = pd.read_csv(r"C:\Users\oriha\Desktop\Serious Sql Excercises\csv\names_with_nulls.csv")
jobs_with_dups_and_nulls = pd.read_csv(r"C:\Users\oriha\Desktop\Serious Sql Excercises\csv\jobs_with_duplicates_and_nulls.csv")
```

```python
# inner join with nulls, on postgress nulls don't join
names_with_nulls.merge(jobs_with_dups_and_nulls, how = "inner")
# oh snap, on pandas nulls do join 
```

```python
# left join with nulls
names_with_nulls.merge(jobs_with_dups_and_nulls, how = "left")
```

```python
# null_jobs as the base
jobs_with_dups_and_nulls.merge(names_with_nulls, how = "left")
```

```python
# left semi with nulls
in_both = names_with_nulls['iid'].isin(jobs_with_dups_and_nulls['iid'])
names_with_nulls[in_both][["iid","first_name"]]
```

```python
# Anti join with nulls
in_both = names_with_nulls['iid'].isin(jobs_with_dups_and_nulls['iid']) == False
names_with_nulls[in_both][["iid","first_name"]]
```

```python
# In
names_with_nulls[names_with_nulls["iid"].isin(jobs_with_dups_and_nulls["iid"].unique())]
```

```python
# Not in
names_with_nulls[names_with_nulls["iid"].isin(jobs_with_dups_and_nulls["iid"].unique()) == False]
```

```python
# Set Operations - Union
pd.concat(objs = [names[names["first_name"] == "Kate"],names[names["first_name"] == "Danny"]])
```

```python
# Set Operations - Union all
pd.concat(objs = [names[names["first_name"] == "Danny"],names[names["first_name"] == "Danny"]])
```

```python
# in pandas all union are union all unless...
pd.concat(objs = [names[names["first_name"] == "Danny"],names[names["first_name"] == "Danny"]]).drop_duplicates()\
                .reset_index(drop = True)
```

```python
# intersect
namedf = names[names["first_name"].str.startswith('K')]
pd.merge(names, namedf, how = "inner")
```

```python
# except
namedf = names[names["first_name"].str.startswith('K') == False]
pd.merge(names, namedf, how = "inner")
```
