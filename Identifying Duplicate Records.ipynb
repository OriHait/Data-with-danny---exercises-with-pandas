```python
# excercise six - What percentage of records are duplicates in the health.user_logs table?
numseries = pd.Series([len(healthdf.drop_duplicates()),len(healthdf) - len(healthdf.drop_duplicates())])
namestype = ['non duplicates','duplicates']
namestype = pd.Series(namestype)
countdf = pd.concat([namestype,numseries],axis = 1).rename(columns={0:'category',1:'count'}).set_index('category')
countdf['percentage'] = countdf['count'] / len(healthdf) 
format_dict = {'percentage': '{:.2%}'}
countdf.style.format(format_dict)
```
